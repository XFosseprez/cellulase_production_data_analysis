## Subject matter opinion and data cleaning 

The run ID suggests that Trichoderma sp. (likely reesei) was used. The parameters pH, DO and steering indicates a liquid state bioreactor was used. As there is a lactose feed indicate, one could assume that fermentation were conducted in a fed-batch mode. 

There are many key pieces of information missing such as the basal medium, source of nitrogen, feeding strategy. Some variable choices are questionable such as the minimal disolved oxygen level rather than the average. Even average values do not indicate if the fungi had been exposed to higher or lower values for a period of time during the fermentation.

The most striking piece of information are the contamination of reactor. In my opinion, such data should not be kept in the dataset as those batches are systematically failed (low yield). By definition, a contamination is unpredictable and affects the yield in an uncontrolled manner. This makes up to 43 % of the whole dataset. Nonetheless, it is going to be removed from the dataset for further analysis.

```python
mask_contamination_N_o = (df_original["Contamination (Y/N)"] == "N")
# Keeping only non-contaminated batches in the dataframe
df_original = df_original[mask_contamination_N_o]
df_original.reset_index(drop= True, inplace= True)

mask_contamination_N = (df["Contamination (Y/N)"] == "N")
# Keeping only non-contaminated batches in the dataframe
df = df[mask_contamination_N]
df.reset_index(drop= True, inplace= True)
```

To further clean the dataset several actions are required.
- the Notes column should be corrected (from "Yeild" to "Yield").
- It is not relevant to indicate steering rate decimals. Rounding this column is required.
- Negative values of disolved oxygen are meaningless. An oxygen probe cannot measure the absence of oxygen. Those values should be replaced by 0.
- Negative values of lactose feed and of cellulase concentration are aberrations. In the 20 first results, there is no such negatives values. Hence it is likely the result of the synthetic generation of those data. For the negative values of feed, the whole line should be removed because it implies a batch mode fermentation and all batch mode fermentation in the original dataset are contaminated batches. For the cellulase concentration, the negative values might be replaced by 0 to keep the data. However, considering the fact that even contaminated batches in the original dataset gave a non-zero cellulase concentration, the whole line are going to be removed for synthetic data row.

```python
# creating a list with all columns containing numerical values
columns_name = df.columns.tolist()
columns_name.remove("Run_ID")
columns_name.remove("Contamination (Y/N)")
columns_name.remove("Notes")
columns_name

#Checking if there are negative values where there shouldn't be any
for n in columns_name:
    negative_values = (df[[n]] < 0)
    total_negative = negative_values.sum()
    print(f"In column  {n}  there are  {total_negative} negative values")
```

```python
# code to correct the "Notes" column
df.rename(columns={"Notes": "Yield level"}, inplace=True)

for n in range(0, df["Yield level"].size):
    #The Upper case format of the string in this column were not standardized, 
    # hence to handle exception every string was lowered
    df.loc[n,"Yield level"] = df.loc[n,"Yield level"].lower()
    if df.loc[n,"Yield level"] == "low yeild":
        df.loc[n,"Yield level"] = "Low"
    if df.loc[n,"Yield level"] == "medium yeild":
        df.loc[n,"Yield level"] = "Medium"
    if df.loc[n,"Yield level"] == "high yeild":
        df.loc[n,"Yield level"] = "High"

df_original.rename(columns={"Notes": "Yield level"}, inplace=True)

for n in range(0, df_original["Yield level"].size):
    #The Upper case format of the string in this column were not standardized, 
    # hence to handle exception every string was lowered
    df_original.loc[n,"Yield level"] = df_original.loc[n,"Yield level"].lower()
    if df_original.loc[n,"Yield level"] == "low yeild":
        df_original.loc[n,"Yield level"] = "Low"
    if df_original.loc[n,"Yield level"] == "medium yeild":
        df_original.loc[n,"Yield level"] = "Medium"
    if df_original.loc[n,"Yield level"] == "high yeild":
        df_original.loc[n,"Yield level"] = "High"
```

```python
df["Max_RPM"] =df["Max_RPM"].round(decimals=0)

for n in range(0, df["Min_DO %"].size):
    if df.loc[n, "Min_DO %"] < 0:
        df.loc[n, "Min_DO %"] = 0

df_original["Max_RPM"] =df_original["Max_RPM"].round(decimals=0)

for n in range(0, df_original["Min_DO %"].size):
    if df_original.loc[n, "Min_DO %"] < 0:
        df_original.loc[n, "Min_DO %"] = 0
```

```python
mask_positive_values = (df["Max FPU/ml"] > 0) & (df["Total feed of Lactose (mL)"] > 0)
mask_negative_values = (df["Max FPU/ml"] <= 0) | (df["Total feed of Lactose (mL)"] <= 0)

# Keeping only rows with positive values of cellulase concentration and positive feed.
df_dropped = df[mask_negative_values]
df = df[mask_positive_values]
df.reset_index(drop= True, inplace= True)
```

The yield is indicated through some categories but it would be interresting to have the actual values. 
The yield should be the sum of the ouptut diveded by the sum of the input.

The lactose feed is clearly an input in the production of cellulase. The lactose itself is used as an inducer of cellulase production and a carbon source for growth.
(Su Yan, Yan Xu & Xiao-Wei Yu, 2021, Bioresources and Bioprocessing volume 8, Article number: 107)
https://bioresourcesbioprocessing.springeropen.com/articles/10.1186/s40643-021-00461-8

The initial input within the volume before feeding is not specified. Furthermore, the initial volume is changed in the actual data for TRICH-17 to TRICH-20. This few changes have been amplified while generating the synthetic data as seen by the standard deviation between the original data (200 mL) and the cleaned dataset (723 mL).

A first option is calculating the yield assuming that only the feed contribute to the cellulase production. It could make sense that the batch medium would not contain lactose to favor biomass production for the batch phase without induction. This yield will be called "Yield (FPU/mL of lactose)". 

A second possibility is to consider both the basal medium and feed as equivalent sources of nutrients. Then the yield is equivalent to the maximal concentration. But it is quiet unlikely as explained before. It would have had been better to know the concentration of nutrients in the basal medium, the concentration in the feed and remaining amount after fermentation to calculate an accurate global yield.

Another important pieces of information missing is the biomass generated. A significant amount of nutrients are used for the cell to exist and multiply. The production of cellulase is, of course, also linked to the amount of biomass. 

```python
# Distribution of the initial volume for the actual fermentation data
df_original["Initial volume (mL)"] = df_original["Total media volume (mL)"]-df_original["Total feed of Lactose (mL)"]
df_original["Initial volume (mL)"].describe()
```

```python
# Distribution of the initial volume for the cleaned dataset
df["Initial volume (mL)"] = df["Total media volume (mL)"]-df["Total feed of Lactose (mL)"]
df["Initial volume (mL)"].describe()
```

```python
df["Yield (FPU/mL of lactose)"] = (df["Total media volume (mL)"]*df["Max FPU/ml"])/df["Total feed of Lactose (mL)"]
df["Yield (FPU/mL of lactose)"] = df["Yield (FPU/mL of lactose)"].round(decimals=1)
df.head()
```

## Modeling yield and concentration

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from scipy.stats import norm, shapiro
```

### Yield (FPU/mL of lactose) linear regression model

```python
# Based on initial visualization, a linear regression model is build to explain yield
lr = LinearRegression()
X = df_original[["Min_DO %","Avg_temp", "Avg_pH", "Max_RPM", "Total feed of Lactose (mL)"]]
Y = df_original[["Yield (FPU/mL of lactose)"]]
lr.fit(X,Y)

Y_hat = lr.predict(X)
R2_score = lr.score(X,Y)
rmse = np.sqrt(mean_squared_error(Y, Y_hat))
residuals = (Y-Y_hat)

print(f"The R^2 of the linear regression of the disolved oxygen against the yield is {round(R2_score, 2)}")
print(f"The RMSE of the linear regression of the disolved oxygen against the yield is {round(rmse,1)}")

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, 
                               figsize=(12, 5), sharey=False)

ax1 = sns.residplot(x= Y_hat, y= residuals, ax=ax1)
ax1.set_ylim(-10, 10)
ax1.set_title('Residual Plot')
ax1.set_xlabel('Predicted Values')
ax1.set_ylabel('Residuals')

#ax2.scatter(x=Y, y=Y_hat, color='green', alpha=0.7)
ax2 = sns.regplot(x=Y, y=Y_hat,color='green', ci=95)
ax2.set_xlabel("Actual Yield")
ax2.set_ylabel("Model Predicted Yield")
ax2.set_title("Model predicted versus actual yield (FPU/mL of lactose)")

plt.tight_layout()
plt.show()
```

```python
# Obtaining more detailed statistical metrics to evaluate the model.
X_with_constant = sm.add_constant(X)
model = sm.OLS(Y, X_with_constant)
results = model.fit()

print(results.summary())
print("\nVariance Inflation Factor (VIF):")

# Checking the Variance Inflation Factor (VIF) for multicollinearity that could jeopardize the model.
vif_data = pd.DataFrame()
vif_data["feature"] = X_with_constant.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_constant.values, i) for i in range(X_with_constant.shape[1])]

print(vif_data)
```

```python
# Visual check of the distribution of residuals agains the normal distribution
sns.kdeplot(residuals)

res_ = np.mean(residuals)
res_std = np.std(residuals)
xmin, xmax = plt.xlim()
x_axis = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x_axis, res_, res_std)
plt.plot(x_axis, pdf, color="red", linestyle="--", label="Normal Distribution (Theoretical)")

# Test statistic to assess if the residuals are Normally distributed.
shapiro_statistic, shapiro_p_value = shapiro(residuals)

print(f"Shapiro-Wilk W-statistic: {shapiro_statistic:.4f}")
print(f"P-value: {shapiro_p_value:.4f}")
```

```python
# Testing the model on the whole dataset

X_syn = df[["Min_DO %","Avg_temp", "Avg_pH", "Max_RPM", "Total feed of Lactose (mL)"]]
Y_syn = df[["Yield (FPU/mL of lactose)"]]
Y_syn_hat = lr.predict(X_syn)
R2_score_syn = lr.score(X_syn,Y_syn)
rmse_syn = np.sqrt(mean_squared_error(Y_syn, Y_syn_hat))

print(f"The R^2 of the linear regression on the synthetic dataset is {round(R2_score_syn, 2)}")
print(f"The RMSE of the linear regression on the synthetic dataset is  {round(rmse_syn,1)}")
```

This model is interresting because it has a good fit with a few factors (R^2 coefficient is good at 0.83).The RMSE seems acceptable (3.3 FPU/mL of lactose) compared to a practicaly significant yield difference (Median of yield and maximum yield in the original dataset are respectively: 13 FPU/mL of lactose and 29 FPU/mL of lactose). Residuals are randomly scattered against the predicted value but they are not normaly distributed. One could argue that there is only 12 individuals in the sample.

When applied on the synthetic dataset, the model does not perform well (R^2 of -0.26, RMSE of 18.6). Because the procedure to generate this dataset was not specified, the model performance should be taken with a grain of salt. The poor performance could be linked to a generation of the synthetic data that did not capture the actual effects of the experimental factors.

## Conclusion and recommendations

### Dataset main issues

First and foremost, there should be an investigation in the laboratory to reduce the batch contamination rate (40 % in the original dataset). Ishikawa diagram should be sketched based on the laboratory 5M analysis and potential root causes should be adressed. Continuous monitoring of the contamination rate would provide a good performance indicator of the preventive actions implemented in the laboratory.

While exploring the dataset, it appears that the synthetic generation of extra data was performed by randomly adding noises. It can be argued that the addition of those noises was made without fully understanding the practical meaning or ground for such dispersion. For instance, the initial volume is treated as random when it should be chosen by the experimenter. Important correlations found in the original dataset disappear through the synthetic generation of the extra data. For those reasons, the synthetic dataset should be considered unfit for analysis. The rest of the discussion will be focused on the original dataset and related analysis.
